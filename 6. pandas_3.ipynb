```python
import pandas as pd
import numpy as np
```

```python
def make_df(cols, ind):
    """Quickly make a DataFrame"""
    data = {j : [str(j) + str(i) for i in ind] for j in cols}
    return pd.DataFrame(data, ind)
```

```python
df1 = make_df('AB', [1, 2])
df2 = make_df('AB', [3, 4])

print(df1); print(df2); print(pd.concat([df1, df2]))
```

```python
df3 = make_df('AB', [0, 1])
df4 = make_df('CD', [0, 1])
print(df3); print(df4); print(pd.concat([df3, df4], axis = 1))
```

```python
df5 = make_df('ABC', [1, 2])
df6 = make_df('BCD', [3, 4])

print(df5); print(df6); print(pd.concat([df5, df6]))
```

```python
print(pd.concat([df5, df6], join = 'inner'))
```

```python
print(df5.append(df6))
```

```python
pop = pd.read_csv('data/state-population.csv')
areas = pd.read_csv('data/state-areas.csv')
abbrevs = pd.read_csv('data/state-abbrevs.csv')
```

Given this information, say we want to compute a relatively straightforward result:
rank US states and territories by their 2010 population density.

```python
print(pop.head(),'\n'); print(areas.head, '\n'); print(abbrevs.head)
```

```python
pop.isnull().sum()
```

```python
areas.isnull().sum()
```

```python
abbrevs.isnull().sum()
```

```python
merged = pd.merge(pop, abbrevs, how = 'outer', left_on = 'state/region', right_on = 'abbreviation')
```

```python
merged
```

```python
merged = merged.drop('abbreviation', 1)
merged.head()
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()].head()
```

```python
merged.loc[merged['state'].isnull(), 'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States Of America'

merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on = 'state', how = 'left')
final
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.loc[final['area (sq. mi)'].isnull(), 'state'].unique()
```

```python
final.dropna(inplace = True)
final
```

```python
final.isnull().any()
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010
```

```python
data2010.set_index('state', inplace = True)
data2010.head()
```

```python
density = data2010['population'] / data2010['area (sq. mi)']
density.head()
```

```python
density.sort_values(ascending = False, inplace = True)
density.head()
```

```python
density.tail()
```
